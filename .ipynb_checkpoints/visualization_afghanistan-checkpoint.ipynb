```python
%matplotlib inline

from random import *
import matplotlib.pyplot as plt
import geopandas as gpd
import pysal as ps
import pandas as pd
import numpy as np
import shapely.ops
from shapely.geometry import Point, LineString, Polygon
from scipy.spatial import Voronoi, voronoi_plot_2d
```

# [URAP] (2) Using "Big" Mobile Phone Data to Visualize Patterns of Mobility in Afghanistan

```python
afg_link = 'district398.shp'
afg = gpd.read_file(afg_link)
```

```python
afg.head()
```

# Step 1: Find the district level map of Afghanistan (shape file). Plot it.

```python
afg.plot(figsize=(18,18))
```

```python
centroids = afg.centroid
centroids.head()
```

# Step 2: Plot centroids of each of the districts.

```python
centroids.plot(figsize=(18,18))
```

```python
base = afg.plot(color='white', edgecolor='black', figsize=(18,18))
centroids.plot(ax=base, marker='o', color='blue', markersize=3);
```

```python
points = []
for point in centroids:
    points += [[point.x, point.y]]
```

# Step 3: From the centroids, perform Voronoi tessellation to convert them back to different polygons (let us call them District Boundaries 2)

```python
voronoi = Voronoi(points)
voronoi_plot_2d(voronoi)
voronoi.vertices
```

```python
lines = [
    shapely.geometry.LineString(voronoi.vertices[line])
    for line in voronoi.ridge_vertices
    if -1 not in line
]
polys = []
for poly in shapely.ops.polygonize(lines):
    polys += [poly]
db2 = afg.copy()
db2 = db2[0:len(polys)]
db2["geometry"] = polys
```

```python
db2.plot(figsize=(18,18))
plt.axis([60, 75, 28, 40])
```

```python
num_districts = len(db2)
num_districts
```

# Step 4: Use a random number generator to assign random numbers between 1000 to 10000 to each of the centroids.

```python
rand_arr = np.array([])
for i in range(num_districts):
    rand_arr = np.append(rand_arr, int(uniform(1000,10000)))
len(rand_arr)
```

```python
db2['val'] = rand_arr
```

# Step 5: Create a heatmap of DB2 using these random numbers. 

```python
db2.plot(column='val',figsize=(18,18),cmap='OrRd')
plt.axis([60, 75, 28, 40])
```

# Step 5a: First split the random numbers into equal bin sizes

```python
pd.cut(rand_arr, 10)
```

# Step 5b: Redo the heatmap by splitting the random numbers into quantiles. 

```python
quantiles = pd.qcut(rand_arr, 4, labels=False)
quantiles
```

```python
db2["val"] = quantiles
db2.plot(column='val',figsize=(18,18),cmap='OrRd')
plt.axis([60, 75, 28, 40])
```
